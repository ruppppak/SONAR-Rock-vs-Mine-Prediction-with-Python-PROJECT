```python
# prompt: import neccessary libraries for sonat rock mines prediction

!pip install pandas numpy scikit-learn matplotlib seaborn
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
 sonar_data= pd.read_csv('sonar_data.csv')
```

```python
sonar_data.head()
```

```python
sonar_data['R'].value_counts()
```

```python
sonar_data.groupby('R').mean()
```

```python
X= sonar_data.drop(columns='R', axis=1)
Y= sonar_data['R']
```

```python
print(X)
print(Y)
```

```python
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.1,stratify=Y,random_state=1)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
print(X_train)
print(Y_train)
```

```python
# Train your model
```

## Training using logistic regression

```python
model = LogisticRegression()
```

```python
# training
model.fit(X_train,Y_train)
```

Find Accuracy of the training Data


```python
X_train_predicition= model.predict(X_train)
training_data_accuracy= accuracy_score(X_train_predicition,Y_train)
```

```python
print('Accuracy on training data:',training_data_accuracy*100)
```

Now doing the same on test data


```python
X_test_predicition= model.predict(X_test)
test_data_accuracy= accuracy_score(X_test_predicition,Y_test)
```

```python
print(test_data_accuracy)
```
